```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
import numpy as np
import pandas as pd
import glob

np.random.seed(1)
tf.random.set_seed(1)
```

```python
subjects = ["bradley", "johnathan", "vishal", "parker", "jake", "joe"]
val_test_types = ["nback", "stroop", "sart"]
electrodes = ["AF3", "F7", "F3", "FC5", "FC6", "F4", "F8", "AF4"]
waves = ["BetaL", "Alpha", "Theta"]
datapath = "E:\\WashU\\Neurotech_Club\\Data\\OneDrive-2021-02-13\\"

count = 0
```

```python
control_count = 0
df_val = []
df_control = []
for subj in subjects:
    subj_files = glob.glob(datapath+"{}*.csv".format(subj))
    for f in subj_files:
        count+=1
        check_files = f.split("\\")[-1]
#         print(temp)
        task_type = f.split("\\")[-1].split("_")[1]

#         print(task_type)
#         print()
#         print(task_type)
        temp = pd.read_csv(f, skiprows=1)

        if task_type == "stroop" or task_type == "sart" or task_type == "posner":
            print(task_type)
            if len(df_val):
                df_val = pd.concat([df_val, temp], axis=0)
            else:
                df_val = temp
        elif task_type == "control" and check_files not in ["parker_control_7.csv", "parker_control_6.csv", "parker_control_4.csv", "johnathan_control_2.csv", "parker_control_2.csv", "parker_control_3.csv", "johnathan_control_1.csv", "parker_control_5.csv", "vishal_control_1.csv"]:
            control_count+=1
            print(task_type)
            print(check_files)
            if len(df_control):
                df_control = pd.concat([df_control, temp], axis=0)
            else:
                df_control = temp
        else:
            None
#             TODO: warning
```

```python
regexp = "Timestamp|"
for electrode in electrodes:
    regexp += "EEG." + electrode + "|" # Get signal from EEG electrode (I'm assuming we will want this for every model; if we do not, add an extra hyperparameter to config)
    for wave in waves:
        regexp += "POW." + electrode + "." + wave + "|"     
regexp = regexp[:-1] # remove the final |

print("regex to pass to dataframe:\n\n" + regexp + "\n\n")
```

```python
if len(waves) > 0: # The headset picks up the waves less frequently than the raw EEG signal, drop these rows. Again, if need more control, please add hyperparam to config
    df_val.dropna(subset=["POW.AF3.BetaL"], inplace=True)
    df_control.dropna(subset=["POW.AF3.BetaL"], inplace=True)

df_val.head()
```

```python
df_control.head()
```

```python
df_val = df_val.filter(regex=regexp)
df_val.set_index("Timestamp", inplace=True)
df_val["Label"] = 1
print("df_task.head():")
df_val.head()
```

```python
df_control = df_control.filter(regex=regexp)
df_control.set_index("Timestamp", inplace=True)
df_control["Label"] = 0
print("df_control.head():")
df_control.head()
```

```python
df_all = pd.concat([df_val, df_control])
df_all = df_all.sample(frac=1.0, random_state=1)
```

```python
df_all.head()
```

```python
df_all.iloc[:,:-1] = (df_all.iloc[:,:-1] - df_all.iloc[:,:-1].mean()) / df_all.iloc[:,:-1].std()
```

```python
X = df_all.iloc[:, :-1].to_numpy()
Y = df_all.iloc[:, -1].to_numpy()
```

```python
def dnn_binary(input_shape,
            hidden_layers=[32, 32]):
    X_input = Input(input_shape)

    X = X_input
    for num_nodes in hidden_layers:
        X = Dense(num_nodes, activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(1e-4),
            bias_regularizer=tf.keras.regularizers.l2(1e-4))(X)
    X = Dense(1, activation="sigmoid")(X)

    model = Model(inputs=X_input, outputs=X)

    return model
model = dnn_binary(X.shape[1],
            hidden_layers=[64,64])
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=.001)
model.compile(optimizer=opt, loss=tf.keras.losses.BinaryCrossentropy(from_logits=False), metrics=[tf.keras.metrics.BinaryAccuracy(threshold=0.5)])
    
```
